Toss & City Insights

```python
import matplotlib.pyplot as plt

df.columns = df.columns.str.strip()  # just to be safe
df['Season'].value_counts().sort_index().plot(
    kind='bar', figsize=(8,4), title='Matches per Season'
)
plt.xlabel('Season')
plt.ylabel('Matches')
plt.show()

```

Season-Wise Analysis

```python
df['winner'].value_counts().head(10).plot(kind='bar', color='orange', title='Top 10 Teams by Wins')
plt.xlabel('Team')
plt.ylabel('Wins')
plt.show()
```

Team Wins Visualization

```python
# Toss winner also match winner
same = (df['toss_winner'] == df['winner']).value_counts(normalize=True)*100
print("Toss winner also won match (%):", same[True])
```
